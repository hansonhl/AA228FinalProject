```python
import numpy as np 
from matplotlib import pyplot as plt 
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
```

```python
def process_file(filename):
    lines = list(filter(lambda x: 'Got average utility' in x, open(filename).readlines()))
    utils = [float(line.replace('Got average utility', '').strip()) for line in lines]
    return np.array(utils)
```

```python
config = 1
asyncs = process_file(f'qmdp_discrete_{config}_async.log')
syncs = process_file(f'qmdp_discrete_{config}_sync.log')
plt.plot(asyncs, label="Async")
plt.plot(syncs, label="Sync")
plt.title(f'Room {config}')
plt.xlabel('Iterations')
plt.ylabel('Utility')
plt.legend()
```
